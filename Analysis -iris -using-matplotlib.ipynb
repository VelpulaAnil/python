# Analysis Iris Flower

```python
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\anilv\Desktop\Iris.csv")
```

```python
datac=data.copy()
```

```python
datac
```

```python
datac.head()
```

```python
datac.drop(["Id"],axis=1,inplace=True)
```

```python
datac.head()
```

```python
datac["Species"].unique()
```

```python
setosa=datac.loc[datac["Species"]=='Iris-setosa']
versicolor=datac.loc[datac["Species"]=='Iris-versicolor']
virginica=datac.loc[datac["Species"]=='Iris-virginica']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(setosa["SepalLengthCm"],setosa["SepalWidthCm"],linestyle=" ",marker="o",color="g",markersize=7,label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["SepalWidthCm"],linestyle=" ",marker="o",color="r",markersize=7,label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["SepalWidthCm"],linestyle=" ",marker="o",color="b",markersize=7,label="virginica")
plt.xlabel("sepallength")
plt.ylabel("sepalwidth")
plt.title("SL VS SW")
plt.legend()
```

```python
plt.plot(setosa["SepalLengthCm"],setosa["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="g",label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="r",label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="b",label="virginica")
plt.xlabel("SepalLength")
plt.ylabel("PetalLength")
plt.title("SL VS PL")
plt.legend()
```

```python
plt.plot(setosa["SepalLengthCm"],setosa["PetalWidthCm"],linestyle=" ",marker="o",markersize=7,color="r",label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["PetalWidthCm"],linestyle=" ",marker="o",markersize=7,color="g",label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["PetalWidthCm"],linestyle=" ",marker="o",markersize=7,color="b",label="virginica")
plt.xlabel("SepalLength")
plt.ylabel("PetalWidth")
plt.title("SL VS PW")
plt.legend()
```

```python
plt.plot(setosa["SepalWidthCm"],setosa["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="r",label="setosa")
plt.plot(versicolor["SepalWidthCm"],versicolor["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="g",label="versicolor")
plt.plot(virginica["SepalWidthCm"],virginica["PetalLengthCm"],linestyle=" ",marker="o",markersize=7,color="b",label="virginica")
plt.xlabel("SepalWidth")
plt.ylabel("PetalLength")
plt.title("SW VS PL")
plt.legend()
```
